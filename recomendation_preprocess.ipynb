```python
import pandas as pd
import numpy as np

from ast import literal_eval
```

```python
df_movies = pd.read_csv('data/movies_all.csv')
df_movies.shape
```

```python
df_movies.info()
```

```python
df_tv = pd.read_csv('data/tv_all.csv')
df_tv.shape
```

```python
df_tv.info()
```

```python
df_movies['current_season'] = 0
df_movies['current_season_Episodes'] = 1

df_movies = df_movies.rename(columns={"genres": "genre"})
df_movies = df_movies.drop(['Revenue','Budget'],axis=1)
```

```python
df_movies.info()
```

```python
df_movies.shape, df_tv.shape
```

```python
df_00 = pd.concat([df_movies, df_tv])
df_00.shape
```

```python
df_00.info()
```

```python
df_00['director'] = df_00['director'].fillna(value=0)
df_00 = df_00.dropna()
```

```python
df_00['genre'] = df_00['genre'].apply(lambda x: literal_eval(x))
df_00['cast'] = df_00['cast'].apply(lambda x: literal_eval(x))
df_00.shape
```

```python
df_00.info()
```

```python
genres_list = []
for movie in df_00.genre.values.tolist():
    for genre in movie:
        if genre not in genres_list:
            genres_list.append(genre)
```

```python
all_genre_list = []
for movie in df_00.genre.values.tolist()[:]:
    genre_list = []
    for genre in genres_list:
       
        if genre in movie:
            genre_list.append(1)
        if genre not in movie:
            genre_list.append(0)
    all_genre_list.append(genre_list)
```

```python
df_generes = pd.DataFrame(all_genre_list,columns=genres_list)
df_generes.shape
```

```python
df_generes.shape, df_00.shape
```

```python
df_01 = df_00.join(df_generes)
df_01.shape
```

```python
cast_list = []
for movie in df_00.cast.values.tolist():
    for cast in movie[:3]:
        if cast not in cast_list:
            cast_list.append(cast)
```

```python
len(cast_list)
```

```python
df_00.cast.values.tolist()[:5]
```

```python
all_cast_list = []
for movie in df_00.cast.values.tolist()[:]:
    acter_list = []
    for cast in cast_list:
        if cast in movie:
            acter_list.append(1)
        if cast not in movie:
            acter_list.append(0)
    all_cast_list.append(acter_list)
```

```python
len(all_cast_list)
```

```python
df_cast = pd.DataFrame(all_cast_list,columns=cast_list)
df_cast.shape
```

```python
df_02['release_date'] = df_02['release_date'].astype('datetime64')
```

```python
df_02['release_date']= pd.Timestamp.today() - df_02['release_date']
df_02['release_date'] = df_02['release_date'].dt.days
```

```python
df_03 = df_02[df_02['release_date'] >0]
df_03.shape
```

```python
df_04 = df_03.reset_index()
```

```python
def time_calculate(run_sample):
    run_amount = 0
    if run_sample[1] == 'h':
        run_amount += 60*int(run_sample[0])
    try:
        if run_sample[-1] == 'm':
            run_amount += int(run_sample[3:-1])
    except ValueError:
        pass
        
    try:
        if run_sample[2] == 'm':
            run_amount += int(run_sample[:2])
    except IndexError:
        pass
    return run_amount
```

```python
df_04['runtime'] = df_04['runtime'].apply(lambda x: time_calculate(x))
```

```python
len(df_04.type.value_counts())
```

```python
df_04.shape
```

```python
df_05 = pd.get_dummies(df_04, columns = ['type','language'], drop_first=True)
df_05.shape
```

```python
df_05 = df_05[df_05.current_season_Episodes != ' Episodes']
df_05.shape
```

```python
def transform_seasons(sample_ep):
    try:
        if sample_ep[-9:] == ' Episodes':
            return int(sample_ep[:-9])
    except TypeError:
            return int(sample_ep)
```

```python
df_05['current_season_Episodes'] = df_05['current_season_Episodes'].apply(lambda x: transform_seasons(x))
```

```python
columns_list_to_drop = ['genre','cast','Keywords','index','current_season']
```

```python
df_06 =df_05.drop(columns_list_to_drop,axis=1)
df_06.shape
```

```python
df_06.to_csv('data/preproced_data.csv',index=False)
```
